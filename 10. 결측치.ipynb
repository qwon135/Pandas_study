## 데이터 제외하기 dropna

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.dropna(inplace=True) # NaN이 있는 데이터는 삭제 4,5번 삭제
df
```

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

- axis : index of columns  (row 를 지울지 column을 지울지 선택)
- how : any or all (any : NaN이 하나라도 있을 때 해당 데이터 삭제 / all : 해당 row or column의 데이터가 모두 NaN일 때 해당 데이터 삭제)

```python
df.dropna(axis='index',how='any') # NaN이 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') # 이번엔 NaN이 하나라도 있는 column 삭제 (how의 기본 값은 any)
```

```python
df['학교']=np.nan # 학교 데이터를 전부 NaN으로 설정 해보자
df
```

```python
df.dropna(axis='columns',how='all') # 데이터 전체가 NaN인 Column 만 삭제
```
