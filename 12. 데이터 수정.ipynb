# 12. 데이터 수정

```python
import pandas as pd

df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

## Colum 수정

```python
df['학교'].replace({'북산고':'상북고','능남고':'료난고'}) # Column값을  바꿀수 있다 (딕셔너리 형태로 정보 입력)
```

```python
df['SW특기']=df['SW특기'].str.lower() # 'SW특기'를 모두 소문자로 변경
df
```

```python
df['학교']=df['학교']+'등학교' # '북산고'를 '북산고등학교'로 변경 , 문자열뒤에 문자열을 + 해준다.
df
```

## Column 추가

```python
df['총합']=df['국어']+df['영어']+df['수학']+df['과학']+df['사회'] # 총합 점수 column을 추가
df
```

```python
df['결과']='Fail' # 결과 Column을 추가하고 초기값을 Fail로 설정
df
```

```python
df.loc[df['총합']>400,'결과']='pass' # 총합이 400보다 크면 결과를 pass로 설정
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합  Column을 삭제
```

```python
df.drop(columns=['국어','영어','수학']) # 국어 수학 영어 Column을 삭제
```

## Row 추가

```python
df.loc['9번']=['이정환','해남고등학교',184,90,90,90,90,90,'Kotlin',450,'Pass'] # 새로운 Row 추가
df
```

## Row 삭제

```python
df.drop(index='4번') # 지원번호 '4번' row 삭제
```

```python
filt=df['수학']<80 # 수학점수가 80점 미만을 거르는 필터 생성
df[filt]
```

```python
df[filt].index# 수학점수 80점 이하인 학생의 index
```

```python
df.drop(index=df[filt].index)
```

## Cell 수정

```python
df.loc['4번','SW특기']='Python' # 4번 학생의 SW특기 데이터를 Python으로 변경
df
```

```python
df.loc['5번',['학교','SW특기']]=['능남고등학교','C'] # 5번 학생의  학교는 능남고, SW특기 데이터를 C로 변경
df
```

## Column 순서 변경

```python
cols=list(df.columns) # cols 변수 생성
cols
```

```python
df=df[[cols[-1]]+cols[0:-1]] #  마지막 column을 가장 앞으로 가져옴
df
```
