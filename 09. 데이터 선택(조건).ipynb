# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd

df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['키'] >=185 # 학생들의 키가 185 이상인지 True/Flase로 알려줌
```

```python
filt= (df['키'] >=185) # 조건을 변수로 만들 수 도 있다.
df[filt] # 필터를 df에 대입하면 해당하는 데이터 정보 가져옴
```

```python
df[-filt] # 변수에 - 를 취해주면 조건에 해당하지 않는 데이터를 보여줌 / 즉, 키가 185 미만인 데이터
```

```python
df.loc[filt,'수학'] # loc 를 활용한 응용, 키가 185이상인 학생의 수학 데이터
```

```python
df.loc[filt,['이름','수학','과학']] # 키가 185 이상인 학생의 이름, 수학, 과학 데이터
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키']>=185) & (df['학교']=='북산고')] # &를 사용하여 키가 185이상 이면서, 북산고를 다니는 학생 데이터 가져옴.
```

### | 또는


```python
df.loc[(df['키']<170)|(df['키']>200)] # |를 사용하여 키가 200보다 크거나 170보다 작은 학생 데이터 가져옴
```

### str 함수
문자열 관리에 사용

```python
filt=df['이름'].str.startswith('송') #  '이름' 데이터에 '송'으로 시작하는 데이터 가져옴
df[filt]
```

```python
filt=df['이름'].str.contains('태') #  '이름'데이터에 태'가 들어가는 데이터
df[filt]
```

```python
df[-filt] # 이름에 태가 들어가지 않는 사람
```

```python
langs=['Python','Java']
filt=df['SW특기'].isin(langs) # 'SW특기'가 lans(Python이거나 Java)에 있는사람
df[filt]
```

```python
df.loc['7번','SW특기'] # 7번의 SW특기는 PYTHON임에도 불구하고 대문자라 불러오지 않는다.
```

```python
langs=['python','java']
filt=df['SW특기'].str.lower().isin(langs) # 소문자로 변환해주면 모두 불러온다.
df[filt]
```

```python
filt=df['SW특기'].str.contains('Java') # 'Java'라는 글자가 들어가는 특기를 가지고 있는 데이터
df[filt]
```

```python
# 보다시피 오류가 뜬다. 오류가 뜨는 이유는 다음과 같다.
```

```python
df['SW특기'].str.lower() # 오류가 뜨는 이유는 'SW특기'의 값이 없어서 NaN으로 뜨는 데이터가 있기 때문이다.
```

```python
df['SW특기'].str.lower().isin(langs) # NaN은 False
```

```python
df['SW특기'].str.contains('Java') # 찍어보면 4,5번이 NaN이다.
```

```python
df['SW특기'].str.contains('Java',na=True) # NaN 데이터를 모두 True로 만들 수 있다.
```

```python
df['SW특기'].str.contains('Java',na=False) # False로 만들 수 있다.
```

```python
filt=df['SW특기'].str.contains('Java',na=False) #  응용하여 NaN을 False로 만들면 정상적으로 출력된다.
df[filt]
```
